```python
#PART 2
import matplotlib.pyplot as plt
# Given dataset
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='green', label='Weight')
plt.title('Scatter Plot of Age vs Height and Weight')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm) / Weight (kg)')
plt.legend()
plt.grid(True)
plt.show()
# Bar Chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, bar_width, label='Height', color='blue')
plt.bar([i + bar_width for i in index], weight, bar_width, label='Weight', color='green')
plt.title('Bar Chart of Height and Weight')
plt.xlabel('People')
plt.ylabel('Height (cm) / Weight (kg)')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.grid(True)
plt.show()
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(height, bins=5, alpha=0.5, label='Height', color='blue')
plt.hist(weight, bins=5, alpha=0.5, label='Weight', color='green')
plt.title('Histogram of Height and Weight Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
